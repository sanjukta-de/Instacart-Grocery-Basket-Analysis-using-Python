## 03. Creating Crosstabs and Bar Charts

```python
# creating a bar chart for the 'department' column

bar_dept=df['department'].value_counts().plot.bar(title='Frequency of Product Orders in each Department')
```

```python
# exporting the above visualization

bar_dept.figure.savefig(os.path.join(path,'04 Analysis','Visualizations','departments.png'),bbox_inches='tight')
```

```python
# creating a bar chart of the 'loyalty_flag'

loyalty_flag=df['loyalty_flag'].value_counts().plot.bar(title='Customers Brand Loyalty')
```

```python
# exporting the above visualization

loyalty_flag.figure.savefig(os.path.join(path,'04 Analysis','Visualizations','loyalty_flag.png'),bbox_inches='tight')
```

```python
# creating a crosstab between 'busiest_period_of_day' and 'loyalty_flag'

loyalty_busiest_period=pd.crosstab(df['busiest_period_of_day'],df['loyalty_flag'])
```

```python
# creating a visualization for the above crosstab

loyalty_busiest_period_viz=loyalty_busiest_period.plot.bar(title='Ordering Time of the Customers',color=['lightsteelblue','lightblue','tab:blue'])
```

```python
# exporting the above visualization

loyalty_busiest_period_viz.figure.savefig(os.path.join(path,'04 Analysis','Visualizations','loyalty_ordering_time.png'),bbox_inches='tight')
```

```python
# creating a crosstab between 'spending_flag' and 'loyalty_flag'

loyalty_spending=pd.crosstab(df['spending_flag'],df['loyalty_flag'])
```

```python
loyalty_spending
```

```python
# creating a visualization of the 'loyalty_spending' crosstab

loyalty_spending_viz=loyalty_spending.plot.bar(title='Spending Habit of the Customers',color=['lightsteelblue','lightblue','tab:blue'])
```

```python
# exporting the above visualization

loyalty_spending_viz.figure.savefig(os.path.join(path,'04 Analysis','Visualizations','loyalty_spending_flag.png'),bbox_inches='tight')
```

```python
# creating a crosstab between 'busiest_days' and 'loyalty_flag'

loyalty_order_days=pd.crosstab(df['busiest_days'],df['loyalty_flag'])
```

```python
loyalty_order_days
```

```python
# creating a visualization for the 'loyalty_order_days' crosstab

loyalty_order_days_viz=loyalty_order_days.plot.bar(title='Ordering Days of the Customers',color=['lightsteelblue','lightblue','tab:blue'])
```

```python
# exporting the above visualization

loyalty_order_days_viz.figure.savefig(os.path.join(path,'04 Analysis','Visualizations','loyalty_order_days.png'),bbox_inches='tight')
```

```python
# creating a crosstab between 'department' and 'region'

region_dept=pd.crosstab(df['department'],df['region'],dropna=False)
```

```python
region_dept= region_dept.sort_values(by=['South'], ascending=True)
```

```python
# creating a visualization for the 'region_dept' crosstab

region_dept_viz=region_dept.plot.barh(title='Regional Spending Habit across Departments', color=['lightsteelblue','cornflowerblue','lightblue','tab:blue'])
```

```python
# exporting the above visualization

region_dept_viz.figure.savefig(os.path.join(path,'04 Analysis','Visualizations','region_dept.png'),bbox_inches='tight')
```
